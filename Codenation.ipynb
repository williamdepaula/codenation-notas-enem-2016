```python
# Importando as bibliotecas a serem usadas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
# Abrindo arquivo de treino
dados_treino = pd.read_csv('train.csv')
```

```python
# Verificando primeiras linhas dos dados
dados_treino.head()
```

```python
# Verificando ultimas linhas dos dados
dados_treino.tail()
```

```python
# Selecionando algumas colunas de interesse
colunas = ['NU_INSCRICAO', 'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_REDACAO', 'NU_NOTA_MT']
dados_treino = dados_treino[colunas]
dados_treino
```

```python
# Verificando se existem valores nulos
dados_treino.isnull().values.any()
```

```python
# Pegando informações de missing
def get_missing_infos(dados_treino):    
    print("# Linhas no dataframe {0}".format(len(dados_treino)))
    print("# Linhas missing NU_INSCRICAO: {0}".format(dados_treino['NU_INSCRICAO'].isnull().sum()))
    print("# Linhas missing NU_NOTA_CN: {0}".format(dados_treino['NU_NOTA_CN'].isnull().sum()))
    print("# Linhas missing NU_NOTA_CH: {0}".format(dados_treino['NU_NOTA_CH'].isnull().sum()))
    print("# Linhas missing NU_NOTA_LC: {0}".format(dados_treino['NU_NOTA_LC'].isnull().sum()))
    print("# Linhas missing NU_NOTA_REDACAO: {0}".format(dados_treino['NU_NOTA_REDACAO'].isnull().sum()))
    print("# Linhas missing NU_NOTA_MT: {0}".format(dados_treino['NU_NOTA_MT'].isnull().sum()))
    
get_missing_infos(dados_treino)
```

```python
# Preenchendo notas missing com zero
dados_treino['NU_NOTA_CN'].fillna(0, inplace=True)
dados_treino['NU_NOTA_CH'].fillna(0, inplace=True)
dados_treino['NU_NOTA_LC'].fillna(0, inplace=True)
dados_treino['NU_NOTA_REDACAO'].fillna(0, inplace=True)
dados_treino['NU_NOTA_MT'].fillna(0, inplace=True)

get_missing_infos(dados_treino)
```

```python
# variavel a ser prevista
atrib_prev = ['NU_NOTA_MT']

# variaveis preditoras
atributos = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_REDACAO']
```

```python
# Criando objetos
X = dados_treino[atributos].values
Y = dados_treino[atrib_prev].values
Y
```

```python
# definindo a taxa de split
split_test_size = 0.30
```

```python
# Criando dados de treino e de teste
X_treino, X_teste, Y_treino, Y_teste = train_test_split(X, Y, test_size = split_test_size, random_state = 42)
```

```python
# Imprimindo os resultados
print("{0:0.2f}% nos dados de treino".format((len(X_treino)/len(dados_treino.index)) * 100))
print("{0:0.2f}% nos dados de teste".format((len(X_teste)/len(dados_treino.index)) * 100))
```

```python
# Importando o módulo de Regressão Linear do scikit-learn
from sklearn.linear_model import LinearRegression
# Criando o modelo
modelo = LinearRegression()
```

```python
# Treinando o modelo
modelo.fit(X_treino, Y_treino)
```

```python
# Aplicando o modelo treinado nos teste
resultado = modelo.predict(X_teste)
resultado = np.around(resultado, 2)

# Setando valores menor do zero para zero
dados_resposta = pd.DataFrame()
dados_resposta = resultado
print(dados_resposta)
dados_resposta[(dados_resposta < 0), ]=0

print(dados_resposta)
```

```python
# Coeficientes
print('Coeficiente: \n', modelo.coef_)

# MSE (mean square error)
print("MSE: %.2f" % np.mean((modelo.predict(X_teste) - Y_teste) ** 2))

# Score de variação: 1 representa predição perfeita
print('Score de variação: %.2f' % modelo.score(X_teste, Y_teste))
```

```python
# Salvando o modelo de teste
import pickle
filename = 'modelo_treinado.sav'
pickle.dump(modelo, open(filename, 'wb'))
```
